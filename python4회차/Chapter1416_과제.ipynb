[문제3]

숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"  
234567 → "23four5six7"  
10203 → "1zerotwozero3"    

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,   
혹은 바뀌지 않고 그대로인 문자열 s가 주어졌을때
s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

---
참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어  
0	zero  
1	one  
2	two  
3	three  
4	four  
5	five  
6	six  
7	seven  
8	eight  
9	nine  

```python
def replace_words_with_numbers(s):
    # 숫자에 대응되는 영단어 사전
    num_ber_word = {
        'zero': '0',
        'one': '1',
        'two': '2',
        'three': '3',
        'four': '4',
        'five': '5',
        'six': '6',
        'seven': '7',
        'eight': '8',
        'nine': '9'
    }
    
    # 영단어를 숫자로 바꾸기 위해 반복
    for word, num in num_ber_word.items():
        s = s.replace(word, num)
    
    return s

# 예시 문자열
s = "77three4one"
result = replace_words_with_numbers(s)
print(result)  # 출력: 77341
```

[문제 4: 여러 손님 주문 처리]

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
while True:                                                                                                                                 #무한 반복 할거임
    fish_type = input("주문할 붕어빵을 입력하세요. 팥붕어빵, 슈크림붕어빵, 초코붕어빵(종료를 원하시면 종료라고 입력해주세요.)")                     #빵맛, 종료 중 입력을 받음
    if fish_type == "종료":  #만약 종료라는 문자열을 받았다면
        break       #break하셈 멈춰. 이 밑에꺼 안 봐도 댐 ㅇㅇ
    bread_count = int(input("주문할 붕어빵 개수를 입력하세요: "))   #근데 그게 아닌 경우엔 갯수를 받으셈
    
    if inventory[fish_type] >= fish_count:      #위와 동일. 재고가 카운트보다 많으면
        inventory[fish_type] -= fish_count      #인벤토리에서 카운트 수만큼 빼셈
        sales[fish_type] += fish_count
        print(f"{fish_type} {fish_count}개를 판매했습니다.")    #그리고 어떤 맛 몇개 팔았는지 출력
    else:       #그 외에 경우는
        print(f"{fish_type} 재고가 없습니다.")      #해당 맛 + 이렇게 출력
    
    print("현재 재고:")     #break를 제외하곤 그 다음엔 이렇게 출력
    for fish, ea in inventory.items():      #인벤토리를 아이템화(키와 값을 동시에 가져옴)시키고 각각 fish엔 키를 ea엔 값을 넣어줌
        print(f"{fish}: {ea}개")            #그리고 출력함
```
